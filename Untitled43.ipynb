```python
# ============================================
# 과제 1. QR 출석체크 시스템
# ============================================

def check_attendance(registered_students, qr_inputs):
    attendance_table = set()
    print("===== QR 출석체크 결과 =====")

    for student_id in qr_inputs:

        # 1. 가짜 입력 검사
        if not student_id.isdigit():
            print(student_id, ":", "가짜 입력")
            continue

        # 2. 미등록 학생 검사
        if student_id not in registered_students:
            print(student_id, ":", "미등록 학생")
            continue

        # 3. 중복 출석 검사
        if student_id in attendance_table:
            print(student_id, ":", "중복 출석")
            continue

        # 4. 정상 출석 처리
        attendance_table.add(student_id)
        print(student_id, ":", "출석 완료")

    print("\n최종 출석 완료 목록")
    print(attendance_table)


# 정상 학생 목록
registered_students = {
    "2024001",
    "2024002",
    "2024003",
    "2024004",
    "2024005"
}

# QR 입력 데이터
qr_inputs = [
    "2024001",
    "2024002",
    "2024003",
    "2024001",
    "FAKE001",
    "2024004",
    "2024002",
    "9999999"
]

check_attendance(registered_students, qr_inputs)


# ============================================
# 과제 2. 닉네임 중복 방지 시스템
# ============================================

def check_nickname(banned_names, nickname_inputs):
    nickname_table = set()

    print("\n===== 닉네임 중복검사 결과 =====")

    for nickname in nickname_inputs:

        # 1. 금지어 닉네임 검사
        if nickname in banned_names:
            print(nickname, ":", "금지어 닉네임")
            continue

        # 2. 중복 닉네임 검사
        if nickname in nickname_table:
            print(nickname, ":", "중복 닉네임")
            continue

        # 3. 사용 가능한 닉네임 저장
        nickname_table.add(nickname)
        print(nickname, ":", "사용 가능")

    print("\n최종 사용 가능 닉네임 목록")
    print(nickname_table)


# 금지어 목록
banned_names = {
    "admin",
    "manager",
    "root"
}

# 닉네임 입력 데이터
nickname_inputs = [
    "lion",
    "apple",
    "tiger",
    "lion",
    "admin",
    "apple",
    "flower"
]

check_nickname(banned_names, nickname_inputs)
```

```python
import cv2
from pyzbar.pyzbar import decode
import requests
import time

# 1. 서버 주소 설정 (코랩에서 실행 중인 서버 주소를 넣으세요)
SERVER_URL = "https://your-colab-url.googleusercontent.com/scan"

# 카메라 켜기
cap = cv2.VideoCapture(0)
used_codes = set() # 중복 방지

print("📸 스캐너가 실행되었습니다. QR코드를 카메라에 비춰주세요.")

while True:
    success, frame = cap.read()
    if not success:
        break

    # 카메라 화면에서 QR코드 찾기
    for code in decode(frame):
        qr_data = code.data.decode('utf-8') # QR 속 내용 읽기

        # 처음 본 QR코드라면 서버로 전송
        if qr_data not in used_codes:
            print(f"🔍 QR 감지: {qr_data}")

            # 서버에 출석 신호 보내기 (QR 데이터가 URL일 경우와 ID일 경우 처리)
            try:
                # QR 데이터가 id=이름 형태라면 서버에 전달
                res = requests.get(f"{SERVER_URL}?id={qr_data}")
                if res.status_code == 200:
                    print(f"✅ {qr_data}님 출석 처리 완료!")
                    used_codes.add(qr_data)
                else:
                    print("❌ 서버 응답 에러")
            except Exception as e:
                print(f"❌ 서버 연결 실패: {e}")

            time.sleep(2) # 잠시 대기

    # 화면 보여주기
    cv2.imshow('QR Scanner', frame)

    # 'q' 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import IPython
from google.colab import output
from flask import Flask, request, jsonify
import threading
import time

# 1. 파이썬 Flask 서버 설정 (백그라운드 실행)
app = Flask(__name__)
attendance_list = []

@app.route('/scan_api')
def scan_api():
    user_id = request.args.get('id', 'Unknown')
    if user_id not in attendance_list:
        attendance_list.append(user_id)
        return jsonify({"status": "success", "message": f"{user_id} 출석 완료!"})
    return jsonify({"status": "already", "message": "이미 출석됨"})

@app.route('/list')
def get_list():
    return jsonify(attendance_list)

# 서버를 별도 쓰레드에서 실행
def run_server():
    app.run(port=5000)

threading.Thread(target=run_server, daemon=True).start()

# 2. 자바스크립트 기반 QR 스캐너 UI
def start_scanner():
    js = """
    async function startQRScanner() {
        // 레이아웃 생성
        const div = document.createElement('div');
        const video = document.createElement('video');
        const canvas = document.createElement('canvas');
        const resultDiv = document.createElement('div');
        resultDiv.style.padding = "10px";
        resultDiv.style.fontSize = "20px";
        resultDiv.innerHTML = "<b>QR코드를 카메라에 비춰주세요...</b>";

        document.body.appendChild(div);
        div.appendChild(video);
        div.appendChild(resultDiv);

        // QR 인식 라이브러리 로드
        const script = document.createElement('script');
        script.src = "https://cdn.jsdelivr.net/npm/jsqr@1.4.0/dist/jsQR.min.js";
        document.head.appendChild(script);

        const stream = await navigator.mediaDevices.getUserMedia({ video: { facingMode: "environment" } });
        video.srcObject = stream;
        video.setAttribute("playsinline", true);
        await video.play();

        const ctx = canvas.getContext("2d");

        async function tick() {
            if (video.readyState === video.HAVE_ENOUGH_DATA) {
                canvas.height = video.videoHeight;
                canvas.width = video.videoWidth;
                ctx.drawImage(video, 0, 0, canvas.width, canvas.height);
                const imageData = ctx.getImageData(0, 0, canvas.width, canvas.height);

                // QR 인식 시도
                const code = jsQR(imageData.data, imageData.width, imageData.height);

                if (code) {
                    resultDiv.innerHTML = "✅ 감지됨: " + code.data;
                    // 서버로 데이터 전송 (API 호출)
                    const res = await fetch('/scan_api?id=' + encodeURIComponent(code.data));
                    const data = await res.json();
                    alert(data.message);

                    // 연속 스캔 방지 (3초 대기)
                    setTimeout(tick, 3000);
                    return;
                }
            }
            requestAnimationFrame(tick);
        }
        script.onload = () => tick();
    }
    startQRScanner();
    """
    display(IPython.display.Javascript(js))

# 실행부
print("🚀 파이썬 서버가 포트 5000에서 시작되었습니다.")
time.sleep(2) # 서버 시작 대기
start_scanner()
```
